```python
import xml.etree.cElementTree as ET
import pandas as pd
```

```python
#this is only 1 of many data files in the dataset, but lets explore it first
tree = ET.ElementTree(file='../data/medsamp2015a.xml')
```

```python
citations = []
topics = []

#this will iterate through all citations
for i, citation in enumerate(tree.iterfind('MedlineCitation')):
    
    #for each citation, we want all the descriptorNames
    topic_list=[]
    for topic in citation.iterfind('MeshHeadingList/MeshHeading/DescriptorName'):
        topic_list.append(topic.text)
        topics.append(topic.text)
    
    citations.append((i, topic_list))
    
```

```python
citations[0]
```

```python
#now we need a co-ocurrence matrix
```

```python
N = len(citations)
print(N)


```

```python
print(len(topics))
print(len(set(topics)))
```

```python
co_df = pd.DataFrame(index=range(0,N), columns=set(topics))
co_df = co_df.fillna(0)
```

```python
co_df.shape
```

```python
co_df.head()
```

```python
citations[0][1]
```

```python
for citation in citations:
    for topic in citation[1]:
        co_df.ix[citation[0]][topic]+=1
```

```python
co_df.describe()
```

```python
co_transpose = co_df.T
```

```python
co_transpose.shape
```

```python
co_occurrence = co_transpose.dot(co_df)
```
